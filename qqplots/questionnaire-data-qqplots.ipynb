```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

measurements_normal = np.random.normal(loc = 20, scale = 5, size=100)

```

```python
import pylab 
import scipy.stats as stats
def myqqplot(measurements):
    stats.probplot(measurements, dist="norm", plot=pylab)
    pylab.show()   
```

```python
myqqplot(measurements_normal)
```

```python
#Stress awareness
wear = [2,4,3,3,5,4,4,5,5,3,3,2]
fitbit = [2,2,2,1,3,5,3,4,2,3,1,3,1]

myqqplot(wear)
```

```python
myqqplot(fitbit)
```

```python
print("watch group:",np.var(wear))
print("phone group:",np.var(fitbit))
```

# Formal Normality Test
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.normaltest.html

D’Agostino and Pearson’s [1], [2] test

```python
# from scipy import stats
```

```python
def isNormal(measures):
    k2, p = stats.normaltest(measures)
    alpha = 1e-2
    print("p = {:g}".format(p))
    if p < alpha:  # null hypothesis: x comes from a normal distribution
        print("The null hypothesis can be rejected, inputs is not normal.")
    else:
        print("The null hypothesis cannot be rejected, inputs is normal.")
    return p >= alpha
```

```python
print("Input is normal:", isNormal(measurements_normal))
```

```python
print("Input is normal:", isNormal(wear))
```

```python
print("Input is normal:", isNormal(fitbit))
```
